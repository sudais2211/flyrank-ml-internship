I plan to use the following fields as features:

impressions_90d — observed search exposure available before the decision.
avg_position — observed average search position available at the decision moment.
content_age_days — age of the content.
days_since_last_update — freshness signal.
word_count — content length.
char_count — content size.
search_volume — search demand context.
competition — competition context.
cpc — commercial search-value context.
content_type — content format/category.
main_intent — search intent context.
Label / Proxy
ctr — used as the observed CTR proxy for identifying optimization opportunities.

For future modeling, the label should ideally represent a future outcome, such as future CTR or CTR improvement. Therefore, current-period CTR should be treated carefully and not assumed to be a causal future label.

Context
client_hash_id — identifies the client grouping.
content_hash_id — identifies the content page.
report_date — identifies the observation date.
client_has_gsc — indicates whether GSC is available.
client_has_ga4 — indicates whether GA4 is available.
gsc_data_available — indicates whether GSC data is available for the row.
ga4_data_available — indicates whether GA4 data is available for the row.
Excluded
clicks_90d — excluded from direct feature use because CTR is already derived from impressions and clicks, creating potential target leakage.
pageviews_90d — excluded because it is downstream behavior rather than a direct search ranking feature.
sessions_90d — excluded because it represents downstream traffic behavior.
users_90d — excluded because it represents downstream user behavior.
engaged_sessions_90d — excluded because it is downstream engagement behavior.
ai_sessions_90d — excluded because it represents downstream AI traffic behavior.
scroll_events_90d — excluded because it is downstream engagement behavior.
days_with_impressions — excluded from the initial model because it overlaps with exposure history.
days_with_sessions — excluded because it represents downstream traffic activity.

```python
# Define the fields used in the data contract

features = [
    "impressions_90d",
    "avg_position",
    "content_age_days",
    "days_since_last_update",
    "word_count",
    "char_count",
    "search_volume",
    "competition",
    "cpc",
    "content_type",
    "main_intent"
]

label = [
    "ctr"
]

context = [
    "client_hash_id",
    "content_hash_id",
    "report_date",
    "client_has_gsc",
    "client_has_ga4",
    "gsc_data_available",
    "ga4_data_available"
]

excluded = [
    "clicks_90d",
    "pageviews_90d",
    "sessions_90d",
    "users_90d",
    "engaged_sessions_90d",
    "ai_sessions_90d",
    "scroll_events_90d",
    "days_with_impressions",
    "days_with_sessions"
]

print("FEATURES:")
for field in features:
    print("-", field)

print("\nLABEL / PROXY:")
for field in label:
    print("-", field)

print("\nCONTEXT:")
for field in context:
    print("-", field)

print("\nEXCLUDED:")
for field in excluded:
    print("-", field)
```

```python
# Check duplicate client-content-day grain on a sample.
# This avoids scanning the entire 78.8M-row dataset.

sample_size = min(10000, len(ds))
sample = ds.select(range(sample_size))

grain_keys = set()
duplicates = 0

for row in sample:
    key = (
        row["client_hash_id"],
        row["content_hash_id"],
        str(row["report_date"])
    )

    if key in grain_keys:
        duplicates += 1
    else:
        grain_keys.add(key)

print("Rows checked:", sample_size)
print("Unique client-content-day keys:", len(grain_keys))
print("Duplicate keys found in sample:", duplicates)

if duplicates == 0:
    print("Result: No duplicate client-content-day keys observed in this sample.")
else:
    print("Result: Duplicate client-content-day keys were observed in this sample.")
```

```python
# Check dates in a sample instead of scanning the entire dataset.

sample_dates = [
    str(row["report_date"])
    for row in sample
    if row["report_date"] is not None
]

print("Sample minimum date:", min(sample_dates))
print("Sample maximum date:", max(sample_dates))

print("\nThe full March 2026 window was verified separately:")
print("2026-03-01 to 2026-03-31")
```

##Data limitations

This dataset supports observed CTR and search-performance analysis, but it cannot establish causal effects.

The history may be unbalanced across clients and content pages, so not every page has the same observation history.
GSC and GA4 availability varies by row, so metrics cannot always be compared across all observations.
Early rows may be GSC-only or have incomplete data availability.
The observation windows can overlap because fields such as 90-day metrics summarize historical periods.
Current CTR is an observed performance measure and should not automatically be treated as a future outcome.
The data does not directly explain why a page received a particular ranking or CTR.
Hash identifiers allow grouping but do not provide real client or content names.
Results should therefore be treated as observed, measured, directional decision-support rather than causal conclusions.
