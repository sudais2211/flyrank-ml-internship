## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

df = pd.read_csv("./data/raw/content_refresh_anonymized.csv")
print("Dataset Shape:", df.shape)
print("\nColumns:", len(df.columns))

df.head()
```

```python
# Create a simple target for optimization opportunity

median_ctr = df["ctr"].median()

df["target_low_ctr"] = (df["ctr"] < median_ctr).astype(int)

print("Median CTR:", round(median_ctr, 4))

df[["content_id", "ctr", "target_low_ctr"]].head(10)
```

```python
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Average Position:", round(df["avg_position"].mean(), 2))
print("Average Impressions:", round(df["impressions_90d"].mean(), 2))

print("\nTarget Distribution:")
print(df["target_low_ctr"].value_counts())
```

# Why ML Beats a Fixed Rule

A fixed rule uses only one feature, such as CTR.

Machine learning can consider multiple features simultaneously, including impressions, search position, and content characteristics, making prioritization more effective.

The model is intended to support SEO decisions rather than replace human judgment.
